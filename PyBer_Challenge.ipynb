### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride =  (pyber_data_df.groupby(["type"])["fare"].sum() / 
                  pyber_data_df.groupby(["type"])["ride_id"].count())


#average_fare_per_ride = [average_fares.map("${:,.2f}".format)]
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = (pyber_data_df.groupby(["type"])["fare"].sum() / 
                           city_data_df.groupby(["type"]).sum()["driver_count"])



#average_fare_per_driver = [average_fares("${:,.2f}".format)]
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({"Total Rides":total_rides,
                                 "Total Drivers":total_drivers,
                                 "Total Fares":total_fares,
                                 "Average Fares by Ride":average_fare_per_ride,
                                 "Average Fares by Driver":average_fare_per_driver})


pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Faress"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fares by Ride"] = pyber_summary_df["Average Fares by Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fares by Driver"] = pyber_summary_df["Average Fares by Driver"].map("${:,.2f}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame

pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

#using 'to_frame' functioun found in documentation at https://pandas.pydata.org/docs/reference/api/pandas.Series.to_frame.html'

fare_by_city_date_df = pyber_data_df.groupby(["type", "date"]).sum().fare.to_frame()
fare_by_city_date_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.

# using documentation found at https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html
#to drop the index into a column

fare_by_city_date_df = fare_by_city_date_df.reset_index()
fare_by_city_date_df.head()
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 

#using documentation found at https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pivot.html
#to pivot 

fare_by_city_date_df = fare_by_city_date_df.pivot(index="date", columns="type", values= "fare")
```

```python
fare_by_city_date_df.head()
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.

dates_fare_df = fare_by_city_date_df.loc["2019-01-01":"2019-04-29"]
dates_fare_df.head()
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.

dates_fare_df.index = pd.to_datetime(dates_fare_df.index)
dates_fare_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
dates_fare_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
#using documentation on resample() found here: 
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html

weekly_data_df = dates_fare_df.resample("W").sum()
weekly_data_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

#using documentation from https://www.dataquest.io/blog/making-538-plots/ for the df.plot() method 
# for 'fivethirtyeight style' graph

# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
weekly_data_df.plot(figsize= (18, 8))
plt.xlabel("Month")
plt.ylabel("Fare (USD)")
plt.title("Total Fares by City Type")
plt.show()
```
